# loc and iloc

```python
data=pd.Series(['a','b','c','d'],index=[1,3,5,7])
print(data)
print(data.loc)
print(data.iloc)
```

```python
data={"Name":['A','B','C','D'],'Age':['25','21','62','40']}
df = pd.DataFrame(data)
print(df.dtypes)
df
```

```python
#astype to convert one datatype into another datatype
df['Age']=df['Age'].astype(int)
print(df.dtypes)
df
```
