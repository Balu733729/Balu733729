```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
import warnings
warnings.filterwarnings('ignore')
```

Line plot

```python
x=[1,2,3,4,5]
y=[10,14,8,16,12]
plt.plot(x,y,marker='s',linestyle = '-.',color='r',label = 'Line')
plt.title('Line Plot Example')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,12,14,8,16]

plt.scatter(x,y, marker = 'o', color = 'r', label = 'Data Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
```

Bar Plot

```python
categories=['A','B','C','D']
values =[15,8,12,10]

plt.bar(categories, values, color = 'r', alpha = 0.4)
plt.title('Bar Plot Example')
plt.xlabel('X-Axis')
plt.ylabel('values')
plt.grid(axis='y')
plt.show()
```

Histogram

```python
data=[1,2,2,3,3,3,4,4,5,5,5,5]

plt.hist(data, bins=5, color='r', alpha=0.5)
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

Pie chart

```python
labels=['A','B','C','D']
sizes=[20,15,25,40]
plt.pie(sizes,labels=labels, autopct='%1.3f%%',startangle=90,colors=['gold','lightgreen','red','yellow'])
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```
